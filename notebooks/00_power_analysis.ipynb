```python
import hashlib
from datetime import UTC, datetime

from forensics.config import get_project_root, settings
from forensics.utils.provenance import compute_config_hash

n_articles = 400  # illustrative; replace with live COUNT from DB in production
d_min = 0.5
summary = (
    f"Illustrative power note: with n≈{n_articles} articles per author, "
    f"Cohen's d≥{d_min} is in a detectable range for rank-based tests at 80% power "
    "(see formal power calculator in analysis module for exact simulations)."
)
print(summary)

root = get_project_root()
pr_path = root / "docs" / "pre_registration.md"
ts = datetime.now(UTC).isoformat()
ch = compute_config_hash(settings)
body = [
    "# Pre-Registration Snapshot",
    "",
    f"Generated (UTC): {ts}",
    "",
    "## Configuration digest",
    f"`config_hash={ch}`",
    "",
    "## Hypotheses",
    "See notebook 00 narrative cells for full tables.",
]
text = "\n".join(body) + "\n"
digest = hashlib.sha256(text.encode()).hexdigest()[:12]
pr_path.write_text(text + f"\n\n**Snapshot hash:** `{digest}`\n", encoding="utf-8")
print("Wrote", pr_path)
```
